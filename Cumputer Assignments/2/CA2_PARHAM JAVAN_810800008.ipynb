```python
# Define a mapping of categories to numerical values
map_sex = {'male': 0, 'female': 1}
map_embark = {'C': 0, 'Q': 1, 'S' : 2}
map_class = {'First': 1, 'Second': 2, 'Third' : 3}
map_who = {'man': 1, 'woman': 2, 'child' : 0}
map_adult_male = {True: 1, False: 0}
map_deck = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4,
                      'F': 5, 'G': 6, 'H': 7, 'I': 8, 'J': 9,
                      'K': 10, 'L': 11, 'M': 12, 'N': 13, 'O': 14,
                      'P': 15, 'Q': 16, 'R': 17, 'S': 18, 'T': 19,
                      'U': 20, 'V': 21, 'W': 22, 'X': 23, 'Y': 24, 'Z': 25}
map_embark_town = {'Cherbourg': 0, 'Southampton': 1, 'Southampton' : 2}
map_alive = {'no' : 0, 'yes' : 1}
map_alone = {True: 1, False: 0}


# Apply the mapping to a specific column using map()
data_frame['sex'] = data_frame['sex'].map(map_sex)
data_frame['embarked'] = data_frame['embarked'].map(map_embark)
data_frame['class'] = data_frame['class'].map(map_class)
data_frame['who'] = data_frame['who'].map(map_who)
data_frame['adult_male'] = data_frame['adult_male'].map(map_adult_male)
data_frame['deck'] = data_frame['deck'].map(map_deck)
data_frame['embark_town'] = data_frame['embark_town'].map(map_embark_town)
data_frame['alive'] = data_frame['alive'].map(map_alive)
data_frame['alone'] = data_frame['alone'].map(map_alone)




# for testing
print(data_frame.head(4))
# data_frame.info()
```

```python
# Step 1: Create a pivot table
pivot_table = data_frame.pivot_table(index=['pclass', 'sex'],
                                     values=['fare', 'survived'],
                                     aggfunc={'fare': 'mean', 'survived': 'mean'})

# Step 2: Plot the average fare
plt.figure(figsize=(10, 6))
pivot_table['fare'].unstack().plot(kind='bar', color=['skyblue', 'salmon', 'green'], alpha=0.7)
plt.title('Average Fare by Class and Sex')
plt.xlabel('Class and Sex')
plt.ylabel('Average Fare')
plt.xticks(rotation=45)
plt.legend(title='Sex')
plt.grid(axis='y')

# Step 3: Plot the survival rate
plt.figure(figsize=(10, 6))
pivot_table['survived'].unstack().plot(kind='bar', color=['skyblue', 'salmon', 'green'], alpha=0.7)
plt.title('Survival Rate by Class and Sex')
plt.xlabel('Class and Sex')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.legend(title='Sex')
plt.grid(axis='y')

plt.show()
```

```python
# Step 1: Group the data by 'class' and 'survived' columns
grouped_data = data_frame.groupby(['pclass', 'survived'])

# Step 2: Calculate the average fare for each group
average_fare = grouped_data['fare'].mean().unstack()

# Step 3: Plot the grouped bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(average_fare))

bar1 = plt.bar(index, average_fare[0], bar_width, label='Did Not Survive', color='salmon', alpha=0.7)
bar2 = plt.bar([i + bar_width for i in index], average_fare[1], bar_width, label='Survived', color='skyblue', alpha=0.7)

plt.xlabel('Class')
plt.ylabel('Average Fare')
plt.title('Average Fare Paid by Passengers, Grouped by Class and Survival Status')
plt.xticks([i + bar_width / 2 for i in index], average_fare.index)
plt.legend()
plt.grid(axis='y')

plt.show()
```

```python

# Load the dataset
df = pd.read_csv('task2.csv')

# 1. Identify and remove duplicates
duplicate_rows = df[df.duplicated()]
df = df.drop_duplicates()

# 2. Identify and handle missing data (NA)
missing_values = df.isnull().sum()
df = df.dropna()

# 3. Standardize salaries to a single currency
currency_frequencies = df['Salary_Currency'].value_counts()
currencies_to_keep = currency_frequencies[currency_frequencies >= 10].index
df = df[df['Salary_Currency'].isin(currencies_to_keep)]

# 4. Convert currencies to USD
c = CurrencyRates()
exchange_rates = {currency: c.get_rate(currency, 'USD') for currency in currencies_to_keep}
df['Salary_in_USD'] = df.apply(lambda row: row['Salary'] * exchange_rates[row['Salary_Currency']], axis=1)
df['Salary_Currency'] = 'USD'

# 5. Identify the Top 10 Most Popular Job Titles
top_job_titles = df['Job_Title'].value_counts().head(10)

# 6. Identify the Top 10 Highest Salaries
highest_salaries = df.sort_values(by='Salary_in_USD', ascending=False).head(10)

# Visualization
fig, axes = plt.subplots(ncols=3, figsize=(18, 5))

# Salary Distribution (Histogram)
axes[0].hist(df['Salary_in_USD'], bins=20, color='skyblue', edgecolor='black')
axes[0].set_title('Salary Distribution')
axes[0].set_xlabel('Salary (USD)')
axes[0].set_ylabel('Frequency')

# Job Title Distribution (Bar Plot)
job_title_counts = df['Job_Title'].value_counts().head(10)
axes[1].bar(job_title_counts.index, job_title_counts.values, color='lightgreen', edgecolor='black')
axes[1].set_title('Top 10 Most Popular Job Titles')
axes[1].set_xlabel('Job Title')
axes[1].set_ylabel('Frequency')
axes[1].tick_params(axis='x', rotation=45)

# Remote Work Ratio Distribution (Violin Plot)
sns.violinplot(x=df['Remote_Ratio'], ax=axes[2], color='salmon')
axes[2].set_title('Remote Work Ratio Distribution (Violin Plot)')
axes[2].set_xlabel('Remote Work Ratio')
axes[2].set_ylabel('Value')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
